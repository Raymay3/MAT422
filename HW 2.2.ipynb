# **2.2.1. Probability axioms**

***Key Mathematical Concepts***



1.   **Sample Space (S)**: The sample space of an experiment is the set of all possible outcomes. For example, when tossing a coin, the sample space is $S =\{Heads,Tails\}$.
2.   **Events**: An event is any subset of the sample space. It can be a single outcome (simple event) or a collection of outcomes (compound event). For example, in rolling a die, the event $A$ of getting an even number is $A = \{2,4,6\}$.
3.   **Probability Distribution**: A function that assigns a probability to each event. For any event $A$, the probability satisfies $0 ≤ P(A) ≤ 1$, and the probability of the entire sample space is 1: $P(S)=1$.
4.   **Basic Properties of Probability**:
      *   *Additive Rule*: For disjoint (mutually exclusive) events $A$ and $B$, $P(A∪B)=P(A)+P(B)$.
      *   *Complement Rule*: $P(A)=1−P(¬A)$, where $¬A$ is the complement of $A$.
      *   *General Addition Rule*: For any events $A$ and $B$, $P(A∪B)=P(A)+P(B)−P(A∩B)$.

5.   **Equal Probability**: If an experiment has $N$ equally likely outcomes, the probability of each outcome is $1/N$. For an event $A$ containing $N(A)$ outcomes, the probability is $P(A)=N(A)/N$.

```python
import itertools

# Example 1: Sample Space and Events for Tossing a Coin Twice
# -----------------------------------------------------------
# Define the sample space for tossing a coin twice
sample_space = list(itertools.product(['Heads', 'Tails'], repeat=2))
print("Sample Space (S):", sample_space)

# Define events
event_A = [outcome for outcome in sample_space if outcome == ('Heads', 'Heads')]
event_B = [outcome for outcome in sample_space if outcome[0] == 'Heads']

print("Event A (Both Heads):", event_A)
print("Event B (First toss is Heads):", event_B)

# Probability of each simple event in the sample space (equal probabilities)
probability_per_outcome = 1 / len(sample_space)
print(f"Probability of each simple event: {probability_per_outcome}")

# Probability of Event A
probability_A = len(event_A) * probability_per_outcome
print(f"Probability of Event A: {probability_A}")

# Probability of Event B
probability_B = len(event_B) * probability_per_outcome
print(f"Probability of Event B: {probability_B}")

# Probability of the complement of Event A
probability_not_A = 1 - probability_A
print(f"Probability of not A: {probability_not_A}")

# Example 2: Rolling a Die
# ------------------------
# Define the sample space for rolling a fair six-sided die
sample_space_die = [1, 2, 3, 4, 5, 6]
print("\nSample Space for Rolling a Die:", sample_space_die)

# Define events
event_C = [outcome for outcome in sample_space_die if outcome % 2 == 0]  # Even number
event_D = [outcome for outcome in sample_space_die if outcome > 3]       # Number > 3

print("Event C (Even numbers):", event_C)
print("Event D (Numbers > 3):", event_D)

# Probability of each simple event in the sample space (equal probabilities)
probability_per_outcome_die = 1 / len(sample_space_die)
print(f"Probability of each simple event (die): {probability_per_outcome_die}")

# Probability of Event C
probability_C = len(event_C) * probability_per_outcome_die
print(f"Probability of Event C: {probability_C}")

# Probability of Event D
probability_D = len(event_D) * probability_per_outcome_die
print(f"Probability of Event D: {probability_D}")

# Probability of C union D
probability_C_union_D = probability_C + probability_D - len(set(event_C) & set(event_D)) * probability_per_outcome_die
print(f"Probability of C ∪ D: {probability_C_union_D}")

# Probability of C intersection D
probability_C_intersection_D = len(set(event_C) & set(event_D)) * probability_per_outcome_die
print(f"Probability of C ∩ D: {probability_C_intersection_D}")

# Probability of the complement of Event C
probability_not_C = 1 - probability_C
print(f"Probability of not C: {probability_not_C}")

# Example 3: Probability Distribution for a Deck of Cards
# --------------------------------------------------------
# Define the sample space for drawing a single card from a standard deck
suits = ['Hearts', 'Diamonds', 'Clubs', 'Spades']
ranks = ['2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King', 'Ace']
deck = [(rank, suit) for rank in ranks for suit in suits]
print("\nSample Space for a Deck of Cards:", deck[:5], "...")  # Show only first 5 for brevity

# Define events
event_E = [card for card in deck if card[1] == 'Hearts']  # Hearts
event_F = [card for card in deck if card[0] in ['Ace', 'King', 'Queen', 'Jack']]  # Face cards

print("Event E (Hearts):", event_E[:5], "...")  # Show only first 5 for brevity
print("Event F (Face cards):", event_F)

# Probability of each simple event in the sample space (equal probabilities)
probability_per_card = 1 / len(deck)
print(f"Probability of each card: {probability_per_card}")

# Probability of Event E
probability_E = len(event_E) * probability_per_card
print(f"Probability of Event E: {probability_E}")

# Probability of Event F
probability_F = len(event_F) * probability_per_card
print(f"Probability of Event F: {probability_F}")

# Probability of E union F
probability_E_union_F = probability_E + probability_F - len(set(event_E) & set(event_F)) * probability_per_card
print(f"Probability of E ∪ F: {probability_E_union_F}")

# Probability of E intersection F
probability_E_intersection_F = len(set(event_E) & set(event_F)) * probability_per_card
print(f"Probability of E ∩ F: {probability_E_intersection_F}")
```

# **2.2.2. Conditional probability**

***Key Mathematical Concepts***

1.   **Conditional Probability**: The probability of event $A$ occurring given that event $B$ has occurred is calculated as:

\
\begin{aligned}
P(A∣B)= \frac{P(A∩B)}{P(B)}
\end{aligned}

\
Here, $P(A∣B)$ is the conditional probability of $A$ given $B$, $P(A∩B)$ is the probability that both $A$ and $B$ occur, and $P(B)$ is the probability of $B$.

2.   **Multiplication Rule**: The probability of the intersection of two events $A$ and $B$ can be found using:

\
\begin{aligned}
P(A∩B)=P(A∣B)⋅P(B)
\end{aligned}

\
3.   **Independence of Events**: Two events $A$ and $B$ are independent if the occurrence of one event does not affect the probability of the other. Mathematically, $A$ and $B$ are independent if:

\
\begin{aligned}
P(A∣B)=P(A)\; or \; P(A∩B)=P(A)⋅P(B)
\end{aligned}

\
4.   **Mutual Independence**: A set of events $A_1, A_2,..., A_n$ are mutually independent if every subset of the events is independent.

```python
import itertools
import random

# Function to calculate probability of an event
def probability(event, sample_space):
    return len(event) / len(sample_space)

# Example 1: Conditional Probability with a Deck of Cards
# -------------------------------------------------------
# Define the sample space for drawing a single card from a standard deck
suits = ['Hearts', 'Diamonds', 'Clubs', 'Spades']
ranks = ['2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King', 'Ace']
deck = [(rank, suit) for rank in ranks for suit in suits]
print("Sample Space (Deck of Cards):", deck[:5], "...")  # Display a few cards for brevity

# Define Event A: Drawing a heart
event_A = [card for card in deck if card[1] == 'Hearts']

# Define Event B: Drawing a face card (Jack, Queen, or King)
event_B = [card for card in deck if card[0] in ['Jack', 'Queen', 'King']]

# Probability of Event A
P_A = probability(event_A, deck)
print(f"Probability of drawing a heart (P(A)): {P_A:.4f}")

# Probability of Event B
P_B = probability(event_B, deck)
print(f"Probability of drawing a face card (P(B)): {P_B:.4f}")

# Probability of Event A ∩ B (drawing a face card that is a heart)
event_A_and_B = [card for card in event_A if card in event_B]
P_A_and_B = probability(event_A_and_B, deck)
print(f"Probability of drawing a face card that is a heart (P(A ∩ B)): {P_A_and_B:.4f}")

# Conditional Probability P(A | B)
P_A_given_B = P_A_and_B / P_B
print(f"Probability of drawing a heart given it's a face card (P(A | B)): {P_A_given_B:.4f}")

# Check Independence: P(A ∩ B) should equal P(A) * P(B)
P_A_times_P_B = P_A * P_B
print(f"Product of P(A) and P(B): {P_A_times_P_B:.4f}")
print(f"Are events A and B independent? {'Yes' if P_A_and_B == P_A_times_P_B else 'No'}")

# Example 2: Conditional Probability with a Die
# ----------------------------------------------
# Define the sample space for rolling a fair six-sided die
sample_space_die = [1, 2, 3, 4, 5, 6]

# Define Event C: Rolling an even number
event_C = [outcome for outcome in sample_space_die if outcome % 2 == 0]

# Define Event D: Rolling a number greater than 3
event_D = [outcome for outcome in sample_space_die if outcome > 3]

# Probability of Event C
P_C = probability(event_C, sample_space_die)
print(f"\nProbability of rolling an even number (P(C)): {P_C:.4f}")

# Probability of Event D
P_D = probability(event_D, sample_space_die)
print(f"Probability of rolling a number greater than 3 (P(D)): {P_D:.4f}")

# Probability of Event C ∩ D (rolling an even number greater than 3)
event_C_and_D = [outcome for outcome in event_C if outcome in event_D]
P_C_and_D = probability(event_C_and_D, sample_space_die)
print(f"Probability of rolling an even number greater than 3 (P(C ∩ D)): {P_C_and_D:.4f}")

# Conditional Probability P(C | D)
P_C_given_D = P_C_and_D / P_D
print(f"Probability of rolling an even number given it's greater than 3 (P(C | D)): {P_C_given_D:.4f}")

# Check Independence: P(C ∩ D) should equal P(C) * P(D)
P_C_times_P_D = P_C * P_D
print(f"Product of P(C) and P(D): {P_C_times_P_D:.4f}")
print(f"Are events C and D independent? {'Yes' if P_C_and_D == P_C_times_P_D else 'No'}")

# Example 3: Mutual Independence with Tossing a Coin
# ---------------------------------------------------
# Define the sample space for tossing a coin three times
sample_space_coin = list(itertools.product(['H', 'T'], repeat=3))
print("\nSample Space (Tossing a Coin 3 times):", sample_space_coin[:5], "...")  # Display a few outcomes for brevity

# Define Event E: First toss is Heads
event_E = [outcome for outcome in sample_space_coin if outcome[0] == 'H']

# Define Event F: Second toss is Tails
event_F = [outcome for outcome in sample_space_coin if outcome[1] == 'T']

# Define Event G: Third toss is Heads
event_G = [outcome for outcome in sample_space_coin if outcome[2] == 'H']

# Probability of Event E
P_E = probability(event_E, sample_space_coin)
print(f"Probability of first toss being Heads (P(E)): {P_E:.4f}")

# Probability of Event F
P_F = probability(event_F, sample_space_coin)
print(f"Probability of second toss being Tails (P(F)): {P_F:.4f}")

# Probability of Event G
P_G = probability(event_G, sample_space_coin)
print(f"Probability of third toss being Heads (P(G)): {P_G:.4f}")

# Probability of Event E ∩ F ∩ G
event_E_and_F_and_G = [outcome for outcome in event_E if outcome in event_F and outcome in event_G]
P_E_and_F_and_G = probability(event_E_and_F_and_G, sample_space_coin)
print(f"Probability of E ∩ F ∩ G: {P_E_and_F_and_G:.4f}")

# Check Mutual Independence: P(E ∩ F ∩ G) should equal P(E) * P(F) * P(G)
P_E_times_P_F_times_P_G = P_E * P_F * P_G
print(f"Product of P(E), P(F), and P(G): {P_E_times_P_F_times_P_G:.4f}")
print(f"Are events E, F, and G mutually independent? {'Yes' if P_E_and_F_and_G == P_E_times_P_F_times_P_G else 'No'}")
```

1.   **Example 1: Conditional Probability with a Deck of Cards**:
  *   *Events A and B*: We define events such as drawing a heart (event $A$) and drawing a face card (event $B$).
  *   *Conditional Probability* $P(A∣B)$: The probability of drawing a heart given that a face card is drawn is calculated.
  *   *Independence Check*: We check if drawing a heart and drawing a face card are independent events using the property $P(A∩B)=P(A)⋅P(B)$.
2.   **Example 2: Conditional Probability with a Die**:
  *   *Events C and D*: We define events like rolling an even number (event $C$) and rolling a number greater than 3 (event $D$).
  *   *Conditional Probability* $P(C∣D)$: The probability of rolling an even number given the number is greater than 3 is calculated.
  *   *Independence Check*: We check if rolling an even number and rolling a number greater than 3 are independent.
3.   **Example 3: Mutual Independence with Tossing a Coin**:
  *   *Events E, F, and G*: We define events based on the outcomes of coin tosses, such as the first toss being heads, the second toss being tails, and the third toss being heads.
  *   *Mutual Independence*: We verify if these events are mutually independent by checking if $P(E∩F∩G)=P(E)⋅P(F)⋅P(G)$.

# **2.2.3. Discrete random variables**

***Key Mathematical Concepts***

1.   **Discrete Random Variables**: A random variable $X$ is discrete if it can take on a countable number of distinct values. Its probability distribution is described by a *probability mass function* (pmf), which gives the probability that $X$ is exactly equal to a given value.
2.   **Probability Mass Function** (**pmf**): The pmf $p(x)$ of a discrete random variable $X$ is defined as:

\
\begin{aligned}
p(x)=P(X=x)
\end{aligned}

\
This function provides the probability of each possible value of $X$.

3.   **Cumulative Distribution Function** (**cdf**): The cdf $F(x)$ of a discrete random variable $X$ is defined as:

\
\begin{aligned}
F(x)=P(X≤x)= ∑_{y≤x} p(y)
\end{aligned}

\
This function gives the probability that $X$ is less than or equal to a given value.

4.   **Expected Value and Variance**:
*   The *expected value* $E(X)$ of a discrete random variable $X$ is a measure of the center of the distribution and is given by:

\
\begin{aligned}
E(X)= ∑_xx⋅p(x)
\end{aligned}


*   The variance $V(X)$ measures the spread of the distribution and is given by:

\
\begin{aligned}
V(X)= ∑_x(x−μ)^2⋅p(x)
\end{aligned}

\
where $μ=E(X)$.

5.   **Bernoulli Distribution**: A random variable $X$ that has only two possible outcomes, 0 (failure) and 1 (success), is called a *Bernoulli random variable*. The pmf of a Bernoulli random variable is:

\
\begin{aligned}
    p(x)=
\begin{cases}
    p,& \text{if } x=1\\
    1-p,  & \text{if } x=0\\
\end{cases}
\end{aligned}

\
where $p$ is the probability of success.

6.   **Binomial Distribution**: A Binomial random variable represents the number of successes in $n$ independent Bernoulli trials. Its pmf is:

\
\begin{aligned}
b(x;n,p)={n\choose x}p^x(1−p)^{n−x} , \;x=0,1,2,...,n
\end{aligned}

\
7.   **Poisson Distribution**: A Poisson random variable represents the number of events occurring in a fixed interval of time or space, with the events occurring independently at a constant rate $μ$. Its pmf is:

\
\begin{aligned}
p(x;μ)= \frac{e^{−μ}μ^x}{x!} , \;x=0,1,2,...
\end{aligned}

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom, poisson

# Function to calculate expected value of a discrete random variable
def expected_value(values, probabilities):
    return np.sum(values * probabilities)

# Function to calculate variance of a discrete random variable
def variance(values, probabilities, mean):
    return np.sum((values - mean) ** 2 * probabilities)

# Example 1: Bernoulli Distribution
# ---------------------------------
p = 0.3  # Probability of success
X = np.array([0, 1])  # Possible values of the Bernoulli random variable
pmf_bernoulli = bernoulli.pmf(X, p)

# Calculate expected value and variance
mean_bernoulli = expected_value(X, pmf_bernoulli)
var_bernoulli = variance(X, pmf_bernoulli, mean_bernoulli)

# Print results for Bernoulli distribution
print(f"Bernoulli Distribution (p = {p}):")
print(f"PMF: {pmf_bernoulli}")
print(f"Expected Value (E[X]): {mean_bernoulli}")
print(f"Variance (Var[X]): {var_bernoulli}")

# Plotting the PMF
plt.figure(figsize=(10, 6))
plt.bar(X, pmf_bernoulli, color='skyblue', alpha=0.7)
plt.title("Bernoulli Distribution PMF")
plt.xlabel("X")
plt.ylabel("P(X)")
plt.xticks(X)
plt.grid(True)
plt.show()

# Example 2: Binomial Distribution
# --------------------------------
n = 10  # Number of trials
p = 0.5  # Probability of success
X_binom = np.arange(0, n+1)  # Possible values of the Binomial random variable
pmf_binom = binom.pmf(X_binom, n, p)

# Calculate expected value and variance
mean_binom = expected_value(X_binom, pmf_binom)
var_binom = variance(X_binom, pmf_binom, mean_binom)

# Print results for Binomial distribution
print(f"\nBinomial Distribution (n = {n}, p = {p}):")
print(f"PMF: {pmf_binom}")
print(f"Expected Value (E[X]): {mean_binom}")
print(f"Variance (Var[X]): {var_binom}")

# Plotting the PMF
plt.figure(figsize=(10, 6))
plt.bar(X_binom, pmf_binom, color='lightgreen', alpha=0.7)
plt.title("Binomial Distribution PMF")
plt.xlabel("X")
plt.ylabel("P(X)")
plt.xticks(X_binom)
plt.grid(True)
plt.show()

# Example 3: Poisson Distribution
# -------------------------------
mu = 4  # Mean rate of occurrence
X_poisson = np.arange(0, 15)  # Possible values of the Poisson random variable
pmf_poisson = poisson.pmf(X_poisson, mu)

# Calculate expected value and variance
mean_poisson = expected_value(X_poisson, pmf_poisson)
var_poisson = variance(X_poisson, pmf_poisson, mean_poisson)

# Print results for Poisson distribution
print(f"\nPoisson Distribution (mu = {mu}):")
print(f"PMF: {pmf_poisson}")
print(f"Expected Value (E[X]): {mean_poisson}")
print(f"Variance (Var[X]): {var_poisson}")

# Plotting the PMF
plt.figure(figsize=(10, 6))
plt.bar(X_poisson, pmf_poisson, color='salmon', alpha=0.7)
plt.title("Poisson Distribution PMF")
plt.xlabel("X")
plt.ylabel("P(X)")
plt.xticks(X_poisson)
plt.grid(True)
plt.show()
```

***Explanation of the Output***

1.   **Bernoulli Distribution**:
  *   *PMF*: The probability mass function for a Bernoulli random variable is plotted, showing the probabilities of 0 (failure) and 1 (success).
  *   *Expected Value & Variance*: The expected value is equal to the probability of success, $E(X)=p$. The variance is calculated as $V(X)=p(1−p)$.
2.   **Binomial Distribution**:
  *   *PMF*: The probability mass function for a Binomial random variable with 10 trials and probability of success 0.5 is plotted.
  *   *Expected Value & Variance*: The expected value is $E(X)=np$, and the variance is $V(X)=np(1−p)$.
3.   **Poisson Distribution**:
  *   *PMF*: The probability mass function for a Poisson random variable with mean rate $μ=4$ is plotted.
  *   *Expected Value & Variance*: The expected value and variance for a Poisson random variable are both equal to $μ$.

# **2.2.4. Continues random variables**

***Probability Density Function (PDF)***

A continuous random variable is described by its *probability density function* (pdf), $f(x)$. The probability that the variable falls within a particular range [a, b] is the area under the curve of $f(x)$ over that range.

**Example 1: Uniform Distribution**

The uniform distribution is the simplest form of a continuous distribution, where $f(x)$ is constant over a range [a, b].

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the uniform distribution PDF
def uniform_pdf(x, a, b):
    if a <= x <= b:
        return 1 / (b - a)
    else:
        return 0

# Generate x values and compute the PDF
a, b = 0, 5
x_values = np.linspace(-1, 6, 1000)
pdf_values = [uniform_pdf(x, a, b) for x in x_values]

# Plot the PDF
plt.figure(figsize=(8, 5))
plt.plot(x_values, pdf_values, label=f'Uniform PDF (a={a}, b={b})')
plt.fill_between(x_values, pdf_values, alpha=0.3)
plt.title('Uniform Distribution PDF')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```

**Explanation**

*   The function `uniform_pdf` defines the uniform distribution over the interval [a, b].
*   The PDF is constant between `a` and `b` and zero elsewhere.

***Expected Value and Variance***

The expected value (mean) $μ$ and variance $σ^2$ of a continuous random variable X are defined as:

\
\begin{aligned}
μ_X = \int\limits_{−∞}^∞ x⋅f(x) \ dx
\end{aligned}

\
\begin{aligned}
σ_X^2 = \int\limits_{−∞}^∞ (x−μ)^2⋅f(x) \ dx
\end{aligned}

\
**Example 2: Exponential Distribution**

The exponential distribution is defined as:

\
\begin{aligned}
    f(x;λ)=
\begin{cases}
    λe^{−λx},& \text{if } x≥0\\
    0,  & \text{if } x<0\\
\end{cases}
\end{aligned}

```python
from scipy.integrate import quad

# Define the exponential distribution PDF
def exponential_pdf(x, lam):
    return lam * np.exp(-lam * x) if x >= 0 else 0

# Parameters
lam = 1

# Expected value (mean) calculation
expected_value = quad(lambda x: x * exponential_pdf(x, lam), 0, np.inf)[0]

# Variance calculation
mean_squared = quad(lambda x: x**2 * exponential_pdf(x, lam), 0, np.inf)[0]
variance = mean_squared - expected_value**2

print(f"Expected Value: {expected_value:.2f}")
print(f"Variance: {variance:.2f}")

# Generate and plot the exponential PDF
x_values = np.linspace(0, 10, 1000)
pdf_values = [exponential_pdf(x, lam) for x in x_values]

plt.figure(figsize=(8, 5))
plt.plot(x_values, pdf_values, label=f'Exponential PDF (λ={lam})')
plt.fill_between(x_values, pdf_values, alpha=0.3)
plt.title('Exponential Distribution PDF')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```

**Explanation**

*   The `quad` function from `scipy.integrate` is used to compute the integrals for expected value and variance.
*   The exponential distribution is commonly used to model the time between events in a Poisson process.

***Exponential Distribution***

The exponential distribution is characterized by its rate parameter $λ$. It has the following properties:

*   Mean: $μ = \frac1λ$
*   Variance: $σ^2 = \frac{1}{λ^2}$

***Normal Distribution***

The normal distribution is perhaps the most important continuous distribution in statistics. It is defined by its mean $μ$ and variance $σ^2$.

\
\begin{aligned}
f(x;μ,σ)= \frac{1}{\sqrt{2πσ^2}}e^{−\frac{(x−μ)^2}{2σ^2}}
\end{aligned}

\
**Example 3: Normal Distribution**

```python
from scipy.stats import norm

# Parameters for the normal distribution
mu, sigma = 0, 1

# Generate x values and compute the PDF
x_values = np.linspace(-4, 4, 1000)
pdf_values = norm.pdf(x_values, mu, sigma)

# Plot the Normal Distribution
plt.figure(figsize=(8, 5))
plt.plot(x_values, pdf_values, label=f'Normal PDF (μ={mu}, σ={sigma})')
plt.fill_between(x_values, pdf_values, alpha=0.3)
plt.title('Normal Distribution PDF')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()

# Compute cumulative probability for a range
a, b = -1, 1
cumulative_prob = norm.cdf(b, mu, sigma) - norm.cdf(a, mu, sigma)
print(f"P({a} <= X <= {b}) = {cumulative_prob:.2f}")
```

**Explanation**

*   The normal distribution has the famous "bell curve" shape.
*   `norm.cdf` is used to calculate the cumulative distribution function, giving the probability that the variable falls within a certain range.
